```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_20newsgroups
```

```python
data = fetch_20newsgroups()
```

```python
categories = data.target_names
categories
```

```python
## Train the data on these categories
train = fetch_20newsgroups(subset='train', categories=categories)

# Testing the data for these categories
test = fetch_20newsgroups(subset='test', categories=categories)
```

```python
print(train.data[5])
```

```python
# Import necessary packages
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline


model = make_pipeline(TfidfVectorizer(), MultinomialNB())

# Traininig the model with training data
model.fit(train.data, train.target)


# creating the labels for the test data
labels = model.predict(test.data)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

cm = confusion_matrix(test.target, labels)
sns.heatmap(cm.T, annot=True, fmt='d', xticklabels=train.target_names, yticklabels=train.target_names)
plt.xlabel('True Label')
plt.ylabel('Predicted Label')
plt.show()
```

```python
# Predict category on the new label 
def predict_category(s, train = train, model =model):
    pred = model.predict([s])
    return train.target_names[pred[0]]
```

```python
predict_category('I like to play with volleyball')
```
